```python
# magic command to display matplotlib plots inline within the ipython notebook webpage
%matplotlib inline

# import necessary modules
import pandas as pd, numpy as np, matplotlib.pyplot as plt
import matplotlib.cm as cm, matplotlib.font_manager as fm
import datetime
import geopandas as gp
from scipy import stats, integrate
#Adjusts what can be displayed when calling a pandas dataframe|
pd.set_option('display.height', 1000)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
HEREstats = pd.read_csv('HEREstatsWMEAN.csv')
#HEREstats.info()
HEREstats.head()
```

```python
R_to_A = gp.read_file('data/ManOWarShape/Richmond_Alumni_StatePlane.shp')
SB_to_T = gp.read_file('data/ManOWarShape/SirBarton_Todds_StatePlane.shp')
T_to_R = gp.read_file('data/ManOWarShape/Todds_Richmond_StatePlane.shp')
R_to_A.head()
```

```python
def LINK_DIR(x):
    for i in range(0,len(x)):
        x.loc[i,'LINK_DIR'] = str(x.loc[i,'LINK_ID'])+ str(x.loc[i,'DIR_TRAVEL'])
    return x
```

```python
LINK_DIR(R_to_A)
R_to_A['LINK_DIR']
```

```python
LINK_DIR(SB_to_T)
SB_to_T['LINK_DIR']
```

```python
LINK_DIR(T_to_R)
T_to_R['LINK_DIR']
```

```python
R_to_A_Join = R_to_A[['LINK_DIR','Shape_Leng','geometry','DIR_TRAVEL']]
R_to_A_Join
```

```python
SB_to_T_Join = SB_to_T[['LINK_DIR','Shape_Leng','geometry','DIR_TRAVEL']]
SB_to_T_Join
```

```python
T_to_R_Join = T_to_R[['LINK_DIR','Shape_Leng','geometry','DIR_TRAVEL']]
T_to_R_Join
```

```python
RA = pd.merge(R_to_A_Join,HEREstats,left_on = 'LINK_DIR', right_on = 'LINK_DIR')
RA.head()
```

```python
TR = pd.merge(T_to_R_Join,HEREstats,left_on = 'LINK_DIR', right_on = 'LINK_DIR')
TR.head()
```

```python
SBT = pd.merge(SB_to_T_Join,HEREstats,left_on = 'LINK_DIR', right_on = 'LINK_DIR')
SBT.head()
```

```python
RA.to_csv('Richmond_AlumniHERE.csv')
SBT.to_csv('SirBarton_ToddsHERE.csv')
TR.to_csv('Todds_RichmondHERE.csv')
```

```python
for i in range(0,len(RA)):
    RA.loc[i,'Weight'] = RA.loc[i,'Shape_Leng']/RA['Shape_Leng'].sum()
RA['Weight']
```

```python
for i in range(0,len(SBT)):
    SBT.loc[i,'Weight'] = SBT.loc[i,'Shape_Leng']/SBT['Shape_Leng'].sum()
SBT['Weight']
```

```python
for i in range(0,len(TR)):
    TR.loc[i,'Weight'] = TR.loc[i,'Shape_Leng']/TR['Shape_Leng'].sum()
TR['Weight']
```

```python
columnsRA = list(RA.columns.values)
columnsRA.remove('LINK_DIR')
columnsRA.remove('Shape_Leng')
columnsRA.remove('geometry')
columnsRA.remove('Weight')
columnsRA.remove('DIR_TRAVEL')
columnsSBT = list(SBT.columns.values)
columnsSBT.remove('LINK_DIR')
columnsSBT.remove('Shape_Leng')
columnsSBT.remove('geometry')
columnsSBT.remove('Weight')
columnsTR = list(TR.columns.values)
columnsTR.remove('LINK_DIR')
columnsTR.remove('Shape_Leng')
columnsTR.remove('geometry')
columnsTR.remove('Weight')
```

```python
for column in columnsRA:
    RA[column] = RA[column] * RA['Weight']
```

```python
for column in columnsSBT:
    SBT[column] = SBT[column] * SBT['Weight']
```

```python
for column in columnsTR:
    TR[column] = TR[column] * TR['Weight']
```
