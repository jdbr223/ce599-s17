```python
#Reads in HEREstats data calculated previously
HEREstats = pd.read_csv('HEREstatsWMEAN.csv')
#HEREstats.info()
HEREstats.head()
```

```python
#Reads shapefiles containing length and directional data for links
R_to_A = gp.read_file('data/ManOWarShape/Richmond_Alumni_StatePlane.shp')
SB_to_T = gp.read_file('data/ManOWarShape/SirBarton_Todds_StatePlane.shp')
T_to_R = gp.read_file('data/ManOWarShape/Todds_Richmond_StatePlane.shp')
R_to_A.head()
```

```python
#Function used to combine link ID with direction of travel to be the same as the HEREstats data
def LINK_DIR(x):
    for i in range(0,len(x)):
        x.loc[i,'LINK_DIR'] = str(x.loc[i,'LINK_ID'])+ str(x.loc[i,'DIR_TRAVEL'])
    return x
```

```python
LINK_DIR(R_to_A)
R_to_A['LINK_DIR']
```

```python
LINK_DIR(SB_to_T)
SB_to_T['LINK_DIR']
```

```python
LINK_DIR(T_to_R)
T_to_R['LINK_DIR']
```

```python
#Drops unneccesary data to prepare for the join
R_to_A_Join = R_to_A[['LINK_DIR','Shape_Leng','geometry','DIR_TRAVEL']]
R_to_A_Join
```

```python
#Drops unneccesary data to prepare for the join
SB_to_T_Join = SB_to_T[['LINK_DIR','Shape_Leng','geometry','DIR_TRAVEL']]
SB_to_T_Join
```

```python
#Drops unneccesary data to prepare for the join|
T_to_R_Join = T_to_R[['LINK_DIR','Shape_Leng','geometry','DIR_TRAVEL']]
T_to_R_Join
```

```python
#Creates a joined table with stats and length info for Richmond to Alumni
RA = pd.merge(R_to_A_Join,HEREstats,left_on = 'LINK_DIR', right_on = 'LINK_DIR')
RA.head()
```

```python
#Creates a joined table with stats and length info for Todds to Richmond
TR = pd.merge(T_to_R_Join,HEREstats,left_on = 'LINK_DIR', right_on = 'LINK_DIR')
TR.head()
```

```python
#Creates a joined table with stats and length info for Sir Barton to Todds
SBT = pd.merge(SB_to_T_Join,HEREstats,left_on = 'LINK_DIR', right_on = 'LINK_DIR')
SBT.head()
```

```python
#Writes combined stats by link to csv's
RA.to_csv('Richmond_AlumniHERE_BothDirections.csv')
SBT.to_csv('SirBarton_ToddsHERE_BothDirections.csv')
TR.to_csv('Todds_RichmondHERE_BothDirections.csv')
```

```python
#Note on Direction:
#T represents the inner circle of ManOWar
#F represents the outer circle of ManOWar
```

```python
#Separates inner and outer circle links
RA_T = RA[RA['DIR_TRAVEL']=='T'].reset_index(drop=True)
RA_F = RA[RA['DIR_TRAVEL']=='F'].reset_index(drop=True)
RA_F
```

```python
#Separates inner and outer circle links
SBT_T = SBT[SBT['DIR_TRAVEL']=='T'].reset_index(drop=True)
SBT_F = SBT[SBT['DIR_TRAVEL']=='F'].reset_index(drop=True)
SBT_F
```

```python
#Separates inner and outer circle links
TR_T = TR[TR['DIR_TRAVEL']=='T'].reset_index(drop=True)
TR_F = TR[TR['DIR_TRAVEL']=='F'].reset_index(drop=True)
TR_T
```

```python
def WeightCalc(x):
    for i in range(0,len(x)):
        x.loc[i,'Weight'] = x.loc[i,'Shape_Leng']/x['Shape_Leng'].sum()
    return x['Weight']
```

```python
WeightCalc(RA_T)
```

```python
WeightCalc(RA_F)
```

```python
WeightCalc(SBT_T)
```

```python
WeightCalc(SBT_F)
```

```python
WeightCalc(TR_T)
```

```python
WeightCalc(TR_F)
```

```python
#Drop columns uneccesary to the calculations
def ColumnDrop(x):
    column = list(x.columns.values)
    column.remove('LINK_DIR')
    column.remove('Shape_Leng')
    column.remove('geometry')
    column.remove('Weight')
    column.remove('DIR_TRAVEL')
    return column
```

```python
ColumnRA_T = ColumnDrop(RA_T)
ColumnRA_F = ColumnDrop(RA_F)
ColumnSBT_T = ColumnDrop(SBT_T)
ColumnSBT_F = ColumnDrop(SBT_F)
ColumnTR_T = ColumnDrop(TR_T)
ColumnTR_F = ColumnDrop(TR_F)
```

```python
#Multiplies weight by all number columns containing stats
def WeightMult(x,y):
    for column in y:
        x[column] = x[column] * x['Weight']
```

```python
WeightMult(RA_T,ColumnRA_T)
WeightMult(RA_F,ColumnRA_F)
WeightMult(SBT_T,ColumnSBT_T)
WeightMult(SBT_F,ColumnSBT_F)
WeightMult(TR_T,ColumnTR_T)
WeightMult(TR_F,ColumnTR_F)
```

```python
#Takes the sum of all columns to yield a segment-wide average to compare to Bluetooth
RA_T_AVG = RA_T.sum()
RA_F_AVG = RA_F.sum()
SBT_T_AVG = SBT_T.sum()
SBT_F_AVG = SBT_F.sum()
TR_T_AVG = TR_T.sum()
TR_F_AVG = TR_F.sum()
```
