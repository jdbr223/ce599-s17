```python
def LINK_DIR(x):
    for i in range(0,len(x)):
        x.loc[i,'LINK_DIR'] = str(x.loc[i,'LINK_ID'])+ str(x.loc[i,'DIR_TRAVEL'])
    return x
```

```python
LINK_DIR(R_to_A)
R_to_A['LINK_DIR']
```

```python
LINK_DIR(SB_to_T)
SB_to_T['LINK_DIR']
```

```python
LINK_DIR(T_to_R)
T_to_R['LINK_DIR']
```
